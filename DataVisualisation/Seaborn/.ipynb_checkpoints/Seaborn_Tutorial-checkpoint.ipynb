## Introduction

- We will use Fatal Police Shootings in the US data sets. 
<br> (https://www.kaggle.com/kwullum/fatal-police-shootings-in-the-us)
- We will first read the data and then make a visualization to find answers to various questions. This will be an example of EDA (exploratory data analysis).
- Our main goal is to make visualization with the Seaborn library.

```python
import numpy as np  # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns

from collections import Counter

import warnings 
warnings.filterwarnings("ignore")

import pathlib
pathlib.Path().absolute()
```

```python
# Reading data

medianHouseholdIncome = pd.read_csv("MedianHouseholdIncome2015.csv", encoding="cp1252")
percentagePeopleBelowPovertyLevel = pd.read_csv("PercentagePeopleBelowPovertyLevel.csv", encoding="cp1252")
percentOver25CompletedHighSchool = pd.read_csv("PercentOver25CompletedHighSchool.csv", encoding="cp1252")
policeKillingsUS = pd.read_csv("PoliceKillingsUS.csv", encoding="cp1252")
shareRaceByCity = pd.read_csv("ShareRaceByCity.csv", encoding="cp1252") 

# I added "encoding="cp1252"" for avoid UTF-8 error. 
```

<p id="2"></p> <br>

## Bar Plot

```python
# Our first question is poverty rate of each state
percentagePeopleBelowPovertyLevel.head()
```

```python
percentagePeopleBelowPovertyLevel.info()
```

```python
# We will check poverty_rate for being sure that all values are okay.
percentagePeopleBelowPovertyLevel.poverty_rate.value_counts() 
# - is unknown values. We need to analysis this data and decide what we do. 
# Our point is learning visualisation with seaborn, so we won't look details. We will change all - with 0.
```

```python
percentagePeopleBelowPovertyLevel.poverty_rate.replace(['-'], 0.0, inplace= True)
```

```python
# poverty_rate's data type is object. We will change it with float.
percentagePeopleBelowPovertyLevel.poverty_rate = percentagePeopleBelowPovertyLevel.poverty_rate.astype(float) 
```

```python
# We use geographic areas for visualisation, so we will find unique state values. 
print(percentagePeopleBelowPovertyLevel["Geographic Area"].unique()) 
print(len(percentagePeopleBelowPovertyLevel["Geographic Area"].unique())) 

# After that we will create area list.
area_list = list(percentagePeopleBelowPovertyLevel["Geographic Area"].unique())
```

```python
# Our question is poverty rate of each state. Let's calculate poverty ratio of each areas.
area_poverty_ratio = []
for i in area_list:
    x = percentagePeopleBelowPovertyLevel[percentagePeopleBelowPovertyLevel["Geographic Area"]==i] # filter area
    area_poverty_rate = sum(x.poverty_rate)/len(x) # find rate of area
    area_poverty_ratio.append(area_poverty_rate) # save rate to the list
    
# creating data for visualisation
data = pd.DataFrame({'area_list': area_list, 'area_poverty_ratio': area_poverty_ratio})
# print(data)

#sorting data and creating new index
new_index = (data['area_poverty_ratio'].sort_values(ascending=False)).index.values  
sorted_data = data.reindex(new_index)
# print(sorted_data)
```

```python
# visualisation
plt.figure(figsize=(15,10)) # create a figure which sizes are 15, 10
ax = sns.barplot(x = sorted_data.area_list, y = sorted_data.area_poverty_ratio)
plt.xticks(rotation=45) # rotate y labels
plt.xlabel('States')
plt.ylabel('Poverty Rate')
plt.title('Poverty Rate Given State')
plt.show()
```

```python
# The second question is most common name or surname of killed people
policeKillingsUS.info()
```

```python
policeKillingsUS.head()
```

```python
policeKillingsUS.name.value_counts()
# There is TK TK name which is meaningless. We will drop them, when we will analysis the data.
```

```python
separate = policeKillingsUS.name[policeKillingsUS.name != 'TK TK'].str.split() # drop TK TK and split name and surname
a,b = zip(*separate) # a is name variable, b is surname variable 
name_list= a+b # tuples added each other

# most common names and their counts
name_count = Counter(name_list)
most_common_names = name_count.most_common(15)
print(most_common_names)
x,y = zip(*most_common_names)
x,y = list(x), list(y) # x is names, y is counts.
```

```python
# Visualisation
plt.figure(figsize=(15,10))
ax = sns.barplot(x = x, y = y, palette= sns.cubehelix_palette(len(x)))
plt.xlabel('Name or surname of killed people')
plt.ylabel('Frequency')
plt.title('Most common 15 Name or Surname of killed people')
plt.show()                 
```

```python
# The third question is high school graduation rate of the population that is older than 25 in states 

percentOver25CompletedHighSchool.head()
```

```python
percentOver25CompletedHighSchool.info()
```

```python
percentOver25CompletedHighSchool.percent_completed_hs.value_counts()
# We will replace '-' with 0
```

```python
percentOver25CompletedHighSchool.percent_completed_hs.replace(['-'], 0.0, inplace=True)
# We will change data type of percent_completed_hs
percentOver25CompletedHighSchool.percent_completed_hs = percentOver25CompletedHighSchool.percent_completed_hs.astype(float)

# We will create area list 
area_list = list(percentOver25CompletedHighSchool['Geographic Area'].unique())

# finding rate of high school graduations
area_highschool = []
for i in  area_list:
    x = percentOver25CompletedHighSchool[percentOver25CompletedHighSchool['Geographic Area']==i]
    area_highschool_rate = sum(x.percent_completed_hs)/len(x)
    area_highschool.append(area_highschool_rate)

data = pd.DataFrame({'area_list':area_list, 'area_highschool': area_highschool})
# print(data)

# sorting
new_index = (data['area_highschool'].sort_values(ascending=True)).index.values
sorted_data2 = data.reindex(new_index)
# print(sorted_data2)
```

```python
# Visualisation
plt.figure(figsize=(15,10))
ax = sns.barplot(x = sorted_data2.area_list, y = sorted_data2.area_highschool)
plt.xticks(rotation=45)
plt.xlabel('States')
plt.ylabel('High School Graduate Rate')
plt.title("Percentage of Given State's Population Above 25 that Has Graduated High School")
plt.show()
```

## Horizontal Bar Plot 

```python
# The fourth question is percentage of state's population 
# according to races that are black,white,native american, asian and hispanic
shareRaceByCity.head()
```

```python
shareRaceByCity.info()
```

```python
shareRaceByCity.replace(['-'],0.0,inplace = True)
shareRaceByCity.replace(['(X)'],0.0,inplace = True)

shareRaceByCity.loc[:,['share_white', 'share_black', 'share_native_american', 'share_asian', 'share_hispanic']] = shareRaceByCity.loc[:,['share_white', 'share_black', 'share_native_american', 'share_asian', 'share_hispanic']].astype(float)

area_list = list(shareRaceByCity['Geographic area'].unique())
share_white = []
share_black = []
share_native_american = []
share_asian = []
share_hispanic = []

for i in area_list:
    x = shareRaceByCity[shareRaceByCity['Geographic area']==i]
    share_white.append(sum(x.share_white)/len(x))
    share_black.append(sum(x.share_black)/len(x))
    share_native_american.append(sum(x. share_native_american)/len(x))
    share_asian.append(sum(x.share_asian)/len(x))
    share_hispanic.append(sum(x.share_hispanic)/len(x))
  

# Visualisation

f, ax = plt.subplots(figsize=(9,15))
sns.barplot(x=share_white, y=area_list, color='green', alpha=0.5, label='White')
sns.barplot(x=share_black, y=area_list, color='blue', alpha=0.7, label='Black')
sns.barplot(x=share_native_american, y=area_list, color='cyan', alpha=0.6, label='Native American')
sns.barplot(x=share_asian, y=area_list, color='yellow', alpha=0.6, label='Asian')
sns.barplot(x=share_hispanic, y=area_list, color='red', alpha=0.6, label='Hispanic')

ax.legend(loc='lower right', frameon= True) # place and frame for chart legend
ax.set(xlabel='Percentage of races', ylabel='States', title="Percentage of State's Population According to Races")
plt.show()
```

## Point Plot 

```python
# The fifth question is comparing high school graduation rate vs Poverty rate of each state 

# Normalisation for comparing
sorted_data['area_poverty_ratio'] = sorted_data['area_poverty_ratio']/max(sorted_data['area_poverty_ratio'])
sorted_data2['area_highschool'] = sorted_data2['area_highschool']/max(sorted_data2['area_highschool'])

# data
data = pd.concat([sorted_data, sorted_data2['area_highschool']], axis=1)
#sorting
data.sort_values('area_poverty_ratio', inplace = True)

# Visualisation
f, ax1 = plt.subplots(figsize= (20,10))
sns.pointplot(x='area_list', y='area_poverty_ratio', data=data, color='lime', alpha=0.8)
sns.pointplot(x='area_list', y='area_highschool', data=data, color='red', alpha=0.8)

plt.text(40, 0.5, 'High school graduate ratio', color='red', fontsize=17, style='italic') 
# 40 and 0.5 express location of text. 
plt.text(40, 0.45, 'Poverty ratio', color='lime', fontsize=17, style='italic')
plt.xlabel('States', fontsize=15, color='blue')
plt.ylabel('Values', fontsize=15, color='blue')
plt.title('High School Graduate  VS  Poverty Rate', fontsize=20, color='blue')
plt.grid()
plt.show()
```

<p id="5"></p> <br>

## Joint Plot 

```python
# The sixth question is comparing high school graduation rate vs Poverty rate of each state

# joint kernel density
# pearsonr= if it is 1, there is positive correlation and if it is, -1 there is negative correlation.
# If it is zero, there is no correlation between variables
# Show the joint distribution using kernel density estimation 

import scipy.stats as stats

g = sns.jointplot(data.area_poverty_ratio, data.area_highschool, kind='kde', size=7)
g.annotate(stats.pearsonr) # showing pearson coefficient
plt.savefig('kde_graph.png')
plt.show()
```

```python
# you can change parameters of joint plot
# kind : { scatter | reg | resid | kde | hex }
# Different usage of parameters but same plot with previous one
g = sns.jointplot("area_poverty_ratio", "area_highschool", data=data, size=5, ratio=3, color='r')
g.annotate(stats.pearsonr)
plt.show()
```

## Pie Plot 

```python
# The seventh question is race rates according in policeKillingsUS data
policeKillingsUS.head()
```

```python
policeKillingsUS.race.head(15)
```

```python
policeKillingsUS.race.value_counts()
```

```python
policeKillingsUS.race.dropna(inplace=True)

labels = policeKillingsUS.race.value_counts().index
colors = ['grey','blue','red','yellow','green','brown']
explode = [0,0,0,0,0,0]
sizes = policeKillingsUS.race.value_counts().values

# Visualisation

plt.figure(figsize=(7,7))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Killed People According to Races',color = 'blue',fontsize = 15)
plt.show()
```

```python
# Same graph with different colors
plt.figure(figsize=(7,7))
plt.pie(sizes, explode=explode, labels=labels, colors=sns.cubehelix_palette(6), autopct='%1.1f%%')
# autopct='%1.1f%%' means show only one decimal place
plt.title('Killed People According to Races',color = 'purple',fontsize = 15)
plt.show()
```

## Heat Map

```python
#correlation map
print(data.corr())
# Same question with sixth that is comparing high school graduation rate vs Poverty rate of each state
f, ax = plt.subplots(figsize=(5,5))
sns.heatmap(data.corr(), annot=True, linewidths=0.5,linecolor='red',fmt= '.1f', ax=ax) 
# annot=True means write corrolation values on graph
# linewidths=0.5 express thickness between squares
# linecolor='red' express color of lines between squares 
# fmt= '.1f' express corrolation values decimal place
# ax=ax means put sns graph into plt.subplots templete
plt.show()
```
