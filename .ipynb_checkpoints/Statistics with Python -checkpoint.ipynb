### Level of Measurements 

Level of measurement or scale of measure is a classification that describes the nature of information within the values assigned to variables. Level of measurements are 
    - nominal 
    - ordinal 
    - interval 
    - ratio

Nominal: 
    - Classification, membership
    - Grouping
    - Central tendancy: Mode
    
Ordinal:
    - Comparison, level
    - Sorting
    - Central tendancy: Median
    - Data can be sorted but the relation between these data can not be explained with mathematical. 
    
Interval:
    - Difference, affinity
    - Yardstick 
    - Central tendancy: Mean, deviation
    - There is no zero point. Like temperature. 
    
Ratio:
    - Magnitude, amount
    - Ratio
    - Central tendancy: Geometric mean, Coefficient of variation
    - There is true zero point.Like age, weight.

### Central Tendency

Mean:The average of the numbers
Median: The "middle" of a sorted list of numbers.
Mode: The value that appears most often.

```python
import numpy as np
from scipy import stats
age = [1,2,3,5,6,7,7,10,12,13]

# Mean 
mean_age = np.mean(age)
print(mean_age)

# Median
median_age = np.median(age)
print(median_age)

# Mode
mode_age = stats.mode(age)
print(mode_age)
```

Mean vs. Median: Mean can be affected by outliers.
For example: The boss will increase salaries if average salary is above 5.

```python
salaries = [1,4,3,2,5,4,2,3,1,500]
print("Mean of salary:", np.mean(salaries))
```

Average salary is 52.5. The boss will think he/she pays too much and not raise the salaries. This is not fair because of person who take 500 salary. Median prevent this unfair situation caused for the outlier.   

```python
print("Median of salary:", np.median(salaries))
```

This median is lower than 5. so the boss will raise salaries.

### Dispersion

Dispersion is a statistical term that describes the size of the distribution of values expected for a particular variable. Dispersion can be measured by several different statistics, such as 
    - range
    - variance
    - standard deviation

```python
import matplotlib.pyplot as plt
y = np.random.uniform(5,8,100)
x1 = np.random.uniform(10,20,100)
x2 = np.random.uniform(0,30,100)
plt.scatter(x1,y,color="black")
plt.scatter(x2,y,color="red")
plt.xlim([-1,31])
plt.ylim([2,11])
plt.xlabel("x")
plt.ylabel("y")
print("x1 mean: {} and median: {}".format(np.mean(x1), np.median(x1)))
print("x2 mean: {} and median: {}".format(np.mean(x2), np.median(x2)))
```

According to above code and graph, we can understand the importance of dispersion. 
x1 and x2 values(black and red dots) have very similar mean and median. But when we look at the graph we can say that these values are very different. How we can find the difference? We can use range, variance or/and standart deviation.

Range: The Range is the difference between the lowest and highest values.

Variance: The variance (σ2) is a measure of how far each value in the data set is from the mean.

Standard Deviation: The Standard Deviation is a measure of how spread out numbers are.

<img src="dispersion.PNG" alt="Drawing" style="width: 600px;"/>

```python
age = [1,2,3,5,6,7,7,10,12,13]

#range 

print("Range: ", max(age)-min(age))

#variance
print("Variance: ", np.var(age))

#standard deviation
print("Std: ", np.std(age))
```

### Quartiles 

Quartiles are the values that divide a list of numbers into quarters.
- The median is the number that is in the middle of sequence.
- The lower quartile(first quartile(Q1)(25%)) is the median in between the smallest number and the median.
- The upper quartile (third quartile(Q3)(75%)) is the median in between the median and the largest number. 
- IQR (Inter quartile range) : Q3-Q1
- Outliers: Values outside between Q1 - 1.5IQR and Q3 + 1.5IQR region

<img src="quartile.PNG" alt="Drawing" style="width: 300px;"/>

### Pearson Corrolation Coefficient

Pearson's correlation coefficient is a statistical measure of the strength of a linear relationship between paired data.
- Division of covarianceby standart deviation of variables.
- The value 
    - r = 1 means a perfect positive correlation 
    - r = -1 means a perfect negataive correlation.
- Requirements for Pearson's correlation coefficient
    - Scale of measurement should be interval or ratio
    - Variables should be approximately normally distributed
    - The association should be linear
    - There should be no outliers in the data

<img src="pearsoncoef.PNG" alt="Drawing" style="width: 1000px;"/>

Let's look at pearson coefficient for area mean and radius mean.

```python
p1 = data.loc[:, ["area_mean", "radius_mean"]].corr(method="pearson")
# Below formula equal to pearson corrolation coefficient
p2 = data.radius_mean.cov(data.area_mean)/(data.radius_mean.std()*data.area_mean.std())
print("Pearson Corrolation: ")
print(p1)
print("Pearson Corrolation: ", p2)
```

### Effect Size

Effect size is a statistical concept that measures the strength of the relationship between two variables on a numeric scale.
You can look at the effect size when comparing any two groups to see how substantially different they are. Effect size emphasises the size of the difference.
- The larger the effect size the stronger the relationship between two variables. 
- The effect size of the population can be known by dividing the two population mean differences by their standard deviation.

We will look at Cohen's effect size. 
- Cohen suggest that if 
    - d(effect size) = 0.2, it is small effect size
    - d(effect size) = 0.5, it is medium effect size
    - d(effect size) = 0.8, it is large effect size
    
<img src="effectsize.PNG" alt="Drawing" style="width: 100px;"/>

```python
#let's compare size of the effect between bening radius mean and malignant radius mean
mean_diff = data_malignant.radius_mean.mean() - data_bening.radius_mean.mean() # m1-m2
var_bening = data_bening.radius_mean.var()
var_malignant = data_malignant.radius_mean.var()
var_pooled = (len(data_bening)*var_malignant + len(data_malignant)*var_bening)/float(len(data_bening)+len(data_malignant))
effect_size = mean_diff / np.sqrt(var_pooled)
print("Effect size: ", effect_size)
```

Effect size 1.9 between bening radius mean and malignant radius mean. That is too big and says that two groups are different from each other.

### Conditional Probability

Conditional probability is the probability of one event occurring with some relationship to one or more other events.
- Conditional Probability: Probability of event A given B
    - P(A|B)
<img src="cond.JPG" alt="Drawing" style="width: 200px;"/>

- For example:

    - Event A is that it is raining outside, and it has a 0.3 (30%) chance of raining today.
    - Event B is that you will need to go outside, and that has a probability of 0.5 (50%).

A conditional probability would look at these two events in relationship with one another, such as the probability that it is both raining and you will need to go outside.
- The formula for conditional probability is:

    - P(B|A) = P(A and B) / P(A)

- which you can also rewrite as:

    - P(B|A) = P(A∩B) / P(A)
    
<p id="18"></p>

### Bayes Theorem

Bayes' theorem describes the probability of an event, based on prior knowledge of conditions that might be related to the event. 
- For example, if the risk of developing health problems is known to increase with age, Bayes’s theorem allows the risk to an individual of a known age to be assessed more accurately than simply assuming that the individual is typical of the population as a whole.

<img src="bayes.JPG" alt="Drawing" style="width: 400px;"/>

- Example of Drug Testing:
    - Suppose, a particular test for whether someone has been using cannabis is 90% sensitive, meaning the true positive rate (TPR)=0.90. Therefore it leads to 90% true positive results (correct identification of drug use) for cannabis users.

    - The test is also 80% specific, meaning true negative rate (TNR)=0.80. Therefore the test correctly identifies 80% of non-use for non-users, but also generates 20% false positives, or false positive rate (FPR)=0.20, for non-users.
    
    - Assuming 0.05 prevalence, meaning 5% of people use cannabis, what is the probability that a random person who tests positive is really a cannabis user?

    - The Positive predictive value (PPV) of a test is proportion of persons who are actually positive out of all those testing positive, and can be calculated from a sample as:
        - PPV = True positive / Tested positive
        
    - If sensitivity, specificity, and prevalence are known, PPV can be calculated using Bayes theorem. Let P(User|Positive) mean "the probability that someone is a cannabis user given that they test positive," which is what is meant by PPV. We can write:
        - P(User|Positive) = P(Positive|User)*P(User) / P(Positive)
        - P(User|Positive) = P(Positive|User)*P(User) / (P(Positive|User)*P(User) + P(Positive|Non-user)*P(Non-user)
        - P(User|Positive) = 0.9*0.05 / (0.9*0.05 + 0.2*0.95)
        - P(User|Positive) = 19%
    - Even if someone tests positive, the probability they are a cannabis user is only 19%, because in this group only 5% of people are users, most positives are false positives coming from the remaining 95%.
    
    - If 1,000 people were tested:
        - 950 are non-users and 190 of them give false positive (0.20 × 950)
        - 50 of them are users and 45 of them give true positive (0.90 × 50)
    - The 1,000 people thus yields 235 positive tests, of which only 45 are genuine drug users, about 19%.

## PROBABILITY DISTRIBUTIONS

A probability distribution is a function that describes the likelihood of obtaining the possible values that a random variable can assume.
- A discrete distribution is a statistical distribution that shows the probabilities of outcomes with finite values.
    - Sum of the individual probabilities must be 1.
    - Discrete probability also can be named as probability mass function.
- A continuous distribution describes the probabilities of the possible values of a continuous random variable.
    - Area under the probability curve must be 1.
    
Probility Distributions:
- Discrete Probability Distributions
    - Uniform Distribution
    - Binominal Distribution
    - Poisson Distribution
- Continuous Probability Distributions
    - Gaussian Distribution

### Uniform Distributions

We can throw a dice and get result of 1,2,3,4,5,6. But we couldn't get 4.78. 
- This distributions have sample space.

```python
# Dice example: 
a = np.random.randint(1,7,60000) # 60000 times throwing dice
print("Sample space: ", np.unique(a))
plt.hist(a, bins=12)
plt.ylabel("Number of outcomes")
plt.xlabel("Possible Outcomes")
plt.show()
```

### Binomial Distributions

A binomial distribution can be thought of as simply the probability of a SUCCESS or FAILURE outcome in an experiment or survey that is repeated multiple times.
- There is only two possible result.
- Examples: Flipping coin or bernoulli trail (failure or success)
- Number of success out of n trials
- Trails are independent.
<img src="bernoille.JPG" alt="Drawing" style="width: 200px;"/>
            - n: number of trial
            - p: probability of success
            - r: number of success

```python
#Head or Tail
# Let's say Head is success
n = 2 # number of trial
p = 0.5 # probability of each trials
s = np.random.binomial(n,p,10000) # 10000 number of trials
weights = np.ones_like(s)/float(len(s))  # Normalization
plt.hist(s, weights=weights)
plt.xlabel("Number of success")
plt.ylabel("Probability")
plt.show()
```

Above graph explain that:
- 0.25 prob 0 success (0 Head, 2 Tails)
- 0.50 prob 1 success (1 Head, 1 Tails)
- 0.25 prob 2 success (2 Heads, 0 Tail)

### Gaussian (Normal) Distributions and Z Score

Gaussian (Normal) Distribution can be named as PDF. Also known as bell shaped distribution. It is symmetrical. Let's look one exapmle.

```python
# IQ of 100000 people
#parameters of normal distribution
mu, sigma = 110, 20 # mean and standard deviation
s = np.random.normal(mu, sigma, 100000)
print("Mean: ", np.mean(s))
print("Standard Deviation: ", np.std(s))
# visualization with histogram
plt.figure(figsize= (10, 7))
plt.hist(s, 100, normed=False) # 100 is bar number
plt.ylabel("Frequency")
plt.xlabel("IQ")
plt.title("Histogram of IQ")
plt.show()
```

- As it can be seen from above graph, most of the people are cumulated near to 110.
- How we can know what percentage of people had IQ score between 80 and 140?
- We will use z-score test to answer the above question.
    - z = (x-mean)/std
    - z1 = (80-120)/20 = -1.5
    - z2 = (140-120)/20 = 1.5
    - This means distance between 80 and mean is 1.5 std and distance between mean and 140 is 1.5 std.
    - If you look at the z table, you will see that 1.5 std correspond to 0.4332.
    
<img src="z_score.JPG" alt="Drawing" style="width: 600px;"/>

    - Let's calculate it with 2 because 1 from 80 to mean and other from mean to 140.
    - 0.4332*2 = 0.8664
    - 86.64% of people has an IQ between 80-140.

### Central Limit Theorem

The central limit theorem states that if you have a population with mean μ and standard deviation σ and take sufficiently large random samples from the population with replacement , then the distribution of the sample means will be approximately normally distributed.
- If we can make a normally distributed sampling, we can make general inferences about the population with them.

```python
# Creating population
x = np.random.random_integers(10, size=100000)
plt.hist(x)
plt.show()
```

```python
# Creating sub group of samples (choosing 10000 samples)

import random
mean_sample = []
for i in range(1000):
    sample = random.randrange(5,10)
    mean_sample.append(np.mean(random.sample(list(x), sample)))
plt.hist(mean_sample, bins=50, color="red")
plt.show()
```

```python
#Showing population and sample in same graph.
plt.hist(x, alpha=0.5, density=True)
plt.hist(mean_sample, bins=50, alpha=0.5, color="red", density=True)
plt.title("Central Limit Theorem")
plt.show()
```

### Hypothesis Testing

Hypothesis testing, a method used to determine the accuracy of a hypothesis within a statistical confidence range. A statistical hypothesis test is a method of statistical inference. Commonly, two statistical data sets are compared, or a data set obtained by sampling is compared against a synthetic data set from an idealized model. An alternative hypothesis is proposed for the statistical-relationship between the two data-sets, and is compared to an idealized null hypothesis that proposes no relationship between these two data-sets. 
- Hypothesis tests are used when determining what outcomes of a study would lead to a rejection of the null hypothesis for a pre-specified level of significance. 
- The process of distinguishing between the null hypothesis and the alternative hypothesis is aided by considering two conceptual types of errors.The first type of error occurs when the null hypothesis is wrongly rejected. The second type of error occurs when the null hypothesis is wrongly not rejected.
       - reject 
       - fail to reject
- Steps: 
    - First we create a null hypothesis and assume it is true. 
    - We determine an alternative hypothesis that we will use when null is rejected. 
    - Then we test the null hypothesis. 
    - We take the data for the test and find the p-value using the test statistics.
    - We compare p-value with level of significance.
    - If p value < level of significance, null hypothesis is rejected.
    - If p value > level of significance, null hypothesis is fail to rejected.
    
- Let's look at the real world problems.

#### Hypothesis Testing Example 1 (Real World Example):
- There is a company produce car parking sensor. This sensor collects data for every 3 millisecond. And its margin of error is 0.6 ms. (Its means sensor collect data between every 2.4ms and 3.6 ms.)
    - <b>mean = 3 ms and sigma = 0.6 ms</b>
- The company want to develop a sensor that takes measurements in less time with 99% confidence. (99% is confidence level.)
    - <b>a = 0.01</b> (for 99% confidence level)(If a = 0.05, confidence level is 95%)
- Engineers developed a new sensor and tested it for 50 times. And this sensor collected data for every 2.8 ms on average.
    - <b>n = 50 and new mean = 2.8 ms</b>
- Question is the new sensor is really faster than the first sensor?
- Let's find the answer with hypothesis testing. 
    - <b>Null Hypothesis :</b> The new sensor is not faster than the first sensor. (Mathematical expression: mean >= 3 ms)
    - <b>Alternative Hypothesis :</b> The new sensor is faster than the first sensor. (Mathematical expression: mean < 3 ms)
    - We should apply <b>left tail test</b> (we will look at alternative hypothesis)
    
<img src="tst_type.JPG" alt="Drawing" style="width: 600px;"/>

    - We will use test statistic for finding p-value. We will use z score. 
    
<img src="z_value.JPG" alt="Drawing" style="width: 200px;"/>

 z = (2.8-3) / (0.6/root(50)) = -2.35
 
 <img src="z_table_eksi.JPG" alt="Drawing" style="width: 600px;"/>
 
 According to z score table, p = 0.0094
 - If p value < 0.01 (level of significance or level of confidence), null hypothesis is rejected.
 - 0.0094 < 00.1 , so null <b>hypothesis is rejected.</b> 
 - This means the new sensor is faster than the first sensor.

### T Distribution

Distributions with the same mean may actually differ in terms of variance.
<img src="t_Test_var.JPG" alt="Drawing" style="width: 600px;"/>
<img src="t_v.JPG" alt="Drawing" style="width: 600px;"/>

```python
s1 = np.array([14.6723, 14.5984, 14.9999, 14.8354, 15.4253, 15.4202, 15.6514, 14.4390, 15.4088,
              14.8781, 14.9393, 15.6427, 14.9631, 14.0379, 14.1098, 15.2318])
print("Mean 1: ", np.mean(s1))
print("Standard Deviation 1:", np.std(s1))
print("Variance 1: ", np.var(s1))
s2 = np.array([15.2365, 15.3005, 15.4983, 15.0371, 14.7239, 14.9746, 15.0381, 15.1866, 15.5914, 
               15.4485, 15.5464, 14.8928, 15.3606, 15.1875, 14.4827, 15.2884])
print("Mean 2: ", np.mean(s2))
print("Standard Deviation 2:", np.std(s2))
print("Variance 2: ", np.var(s2))

# visualization with histogram
import seaborn as sns
sns.kdeplot(s1)
sns.kdeplot(s2)
plt.show()
```

```python
t_val = np.abs(np.mean(s1)- np.mean(s2)) / np.sqrt((np.var(s1)/len(s1))+(np.var(s1)/len(s1))) 
# abs is absolute value (mutlak deger)
print("t value: ", t_val)
```

<b>Null hypothesis:</b> There is no statistical difference between these two distributions.
- If t value < critical value, this is fail to reject null hypothesis.
- If t value > critical value, this is reject null hypothesis.

### What is ANOVA?

Analysis of variance (ANOVA) is a collection of statistical models and their associated estimation procedures (such as the "variation" among and between groups) used to analyze the differences among group means in a sample.
- Anova answers the question whether the groups are the same.
- Let's say we want to learn difference of stress levels of students at primary, secondary school and university.
    - Null hypothesis: Stress levels are same.
    - Stress level means and standard deviations:
        - primary: mean= 47 and std= 10
        - secondary: mean= 48 and std= 15
        - university: mean= 55 and std= 5
        - Total mean= 50
    - ANOVA analize two variance: 
        - Variance between groups
        - Variance within groups 
    - F value is ratio of these variance. (Variance between groups)/(Variance within groups) 

```python
primary = np.array([51.3637, 44.9694, 49.4364, 45.8458, 45.7667,
                   56.0403, 60.8516, 39.1679, 36.9013, 43.5808])
secondary = np.array([56.6567, 55.9272, 42.3243, 50.1913, 48.9178,
                      48.1159, 50.9129, 47.4613, 42.7694, 36.8673])
university = np.array([60.0360, 56.9473, 57.7702, 47.2985, 54.2155,
                      57.7400, 50.9241, 53.4777, 55.6296, 59.4298])

print("Primary mean: ", np.mean(primary))
print("Secondary mean: ", np.mean(secondary))
print("University mean: ", np.mean(university))
total_mean = (np.mean(primary)+np.mean(secondary)+np.mean(university))/3
print("Total mean: ", np.mean(total_mean))

#visualization with pdf
sns.kdeplot(primary)
sns.kdeplot(secondary)
sns.kdeplot(university)
plt.show()
```

```python
# Finding f value: statistic in the result = f value
stats.f_oneway(primary, secondary, university)
```

## CHI-SQUARE ANALYSIS <p id="39"></p> 

### What is Chi-Square Analysis?

A chi-square test for independence compares two variables in a contingency table to see if they are related. 
- Example: A coin flipped 10 times and got the result of 9 heads and 1 tail. Is it luck or is this coin rigged?
    - Null hypothesis: It is possible that the getting result of 9 heads and 1 tail with a 95% confidence. (Money is not rigged.)
<img src="chi.JPG" alt="Drawing" style="width: 200px;"/>

- For heads: 
    - expected frequency = 5
    - observed frequency = 9
- For tails:
    - expected frequency = 5
    - observed frequency = 1
- Chi square value = ((9-5)**2/5)+((1-5)**2/5) = 6.4
- Degrees of freedom = number of possible outcomes - 1
- Degrees of freedom = 2-1 = 1 

<img src="chi_table.JPG" alt="Drawing" style="width: 600px;"/>


According to table above, critical value = 3.841

- 6.4 > 3.8, so reject null hypothesis. The coin is rigged.

<b>Another example of Chi square testing</b>
- There are 7 computer and these are crashed several times given below.
- Computers are independent. And their probabilities of failure is the same.
- Observations: computer1 (5), computer2 (7),computer3 (9), computer4 (4), computer5 (1), computer6 (10), computer7 (6)
    - Null hypothesis: The observation results are normal with 95% probability.(All computers are same.)
    - Total crash = 42
    - Expected value = 42/6 = 6
    - Degrees of freedom = 7-1 = 6

```python
observation = np.array([5,7,9,4,1,10,6])
print("Total crash: ", np.sum(observation))
expected = np.sum(observation)/len(observation)
print("Expected: ", expected)
chi_value = np.sum((observation-expected)**2/expected)
print("Chi square value: ", chi_value)
```

```python
from scipy.stats import chi2
print("Critical value: ", chi2.isf(0.05, 6))
```

9.33 (Chi square) < 12.59 (Critical value) , so fail to reject null hypothesis. Computers are the same.
